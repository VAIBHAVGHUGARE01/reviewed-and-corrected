**K-Means**

```python
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
```

```python
# Select the features for clustering
# Replace 'feature1', 'feature2', ... with the actual column names you want to use
features = Crime_x[['Murder',	'Assault',	'UrbanPop',	'Rape']]
```

```python
# Specify the number of clusters (k) - you need to decide on this based on your problem
k = 4
```

```python
# Initialize the KMeans model
kmeans = KMeans(n_clusters=k,init = 'k-means++', random_state=42)
k_means = kmeans.fit_predict(Crime_x)
```

```python
k_means
```

```python
# Fit the model to your crime data
kmeans.fit(features)
```

```python
# Add a new column to your DataFrame to store the cluster labels
Crime_x['Cluster'] = kmeans.labels_
```

```python
# Print the count of data points in each cluster
print(Crime_x['Cluster'].value_counts())
```

```python
# Visualize the clusters (for 2D data, you can modify this for more dimensions)
plt.scatter(features['Murder'], features['UrbanPop'], c=Crime_x['Cluster'], cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', label='Centroids')
plt.title('K-Means Clustering of Crime Data')
plt.xlabel('Murder')
plt.ylabel('UrbanPop')
plt.legend()
plt.show()
```

```python
plt.scatter(features['Rape'], features['UrbanPop'], c=Crime_x['Cluster'], cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', label='Centroids')
plt.title('K-Means Clustering of Crime Data')
plt.xlabel('Rape')
plt.ylabel('UrbanPop')
plt.legend()
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
# Compute the linkage matrix
linked = linkage(features, 'ward')
```

```python
k_values = range(1, 11)
```

```python
# Fit k-means models for each value of k
inertia_values = []
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(features)
    inertia_values.append(kmeans.inertia_)
```

```python
# Plot the elbow method
plt.plot(k_values, inertia_values, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Within-Cluster Sum of Squares (WCSS)')
plt.show()
```
