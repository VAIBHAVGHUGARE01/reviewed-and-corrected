```python
#Normalizing our data
std=StandardScaler()
Wine_normal=std.fit_transform(Wine)
Wine_normal
```

```python
Wine_normal.shape
```

**Creating PCA Modelling**

```python
#Create Instant of PCA
pca=PCA(n_components=13)
pca_components=pca.fit_transform(Wine_normal)
```

```python
#The ammount of varience That each PcA explained
pca.explained_variance_
#It means that the principle component 5.5672 units of variance in our data.
#High variance value indicates that the corresponding principle component captures a significant amount of information.
```

```python
var=pca.explained_variance_ratio_ #This is represented in percentage
```

```python
# Creating a dataframe featuring the two Principal components that we acquired through PCA.
PCA_Winedataset = pd.DataFrame(data = pca_components, columns = ['component1', 'component2', 'component3', 'component4',
                                                                  'component5', 'component6','component7', 'component8', 'component9',
                                                                 'component10', 'component11', 'component12', 'component13'] )
PCA_Winedataset
```

```python
# Cumulative variance
var1 = np.cumsum(np.round(var,decimals = 4)*100)
var1  #It shows the accumulation of variance for each principle component no.It describes the variance of each principle component
```

```python
pca.components_
```

```python
Wine.columns
```

```python
# Variance plot for PCA components obtained
plt.plot(var1,color="blue")
```

**Use first 3 principal components for further computation, Extracting the three features from PCA_dataset**

```python
pca_component1 = PCA_Winedataset['component1']
pca_component2 = PCA_Winedataset['component2']
pca_component3 = PCA_Winedataset['component3']
```

```python
PCA_Wine=pd.concat([PCA_Winedataset['component1'],PCA_Winedataset['component2'],PCA_Winedataset['component3']],axis=1)
PCA_Wine
```

```python
# Visualizing the results of the 3D PCA.
from mpl_toolkits.mplot3d import Axes3D
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.scatter(
    xs=pca_component1,
    ys=pca_component2,
    zs=pca_component3,
)
ax.set_xlabel("X Label")
ax.set_ylabel("Y Label")
ax.set_zlabel("Z Label")
plt.show()
```

```python
#  Normalizing Dataset
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
pca_Wine_normal = scaler.fit_transform(PCA_Wine)
print(pca_Wine_normal)
```

```python
# Creating clusters
from sklearn.cluster import AgglomerativeClustering
H_clusters=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
H_clusters
```

**K-MEANS Clustering**

```python
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(pca_Wine_normal)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#Build Cluster algorithm

KM_clusters = KMeans(3, random_state=42)
KM_clusters.fit(pca_Wine_normal)
```

```python
y=pd.DataFrame(KM_clusters.fit_predict(pca_Wine_normal),columns=['clusterid_Kmeans'])
y['clusterid_Kmeans'].value_counts() # Its create Three clusters one cluster having 67 ,second cluster having 49 and third 62
```

**Preparing Actual Vs. Predicted Clusering Data**

```python
wine_class = Wine['Type']
wine_class = pd.Series(wine_class)
```

```python
clustersid_HC = H_clusters.labels_
clustersid_HC = pd.Series(clustersid_HC)
```

```python
clusterid_Kmeans = KM_clusters.labels_
clusterid_Kmeans = pd.Series(clusterid_Kmeans)
```

```python
pred_Wine = pd.concat([wine_class, clustersid_HC, clusterid_Kmeans],axis = 1)
pred_Wine
#Interpretation- For Type 1 Wine, in one cluster, H-Clustering 60 and in KM-Clustering 62 Datapoint. Like that another type wine clusters occurs..
#Total Three Cluster in H-Clustering and Three in KM clustering
```
